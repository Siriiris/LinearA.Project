```python
# Rename the data set to be easy to use
df.columns = ['month', 'wind', 'solar', 'geothermal', 'wood','otherbiomass']
```

```python
df.head()
```

```python
# Set the data from the past to present 
Rearrange = df.sort_values(by="month")
```

```python
Rearrange
```

```python
#Try to visualize the data
plt.figure(figsize=(20,15))
plt.title('The renewable energy is produced over the time in the US')
plt.plot(Rearrange.month,Rearrange.wind, label= 'Wind')
plt.plot(Rearrange.month,Rearrange.solar, label= 'Solar')
plt.plot(Rearrange.month,Rearrange.geothermal, label= 'Geothermal')
plt.plot(Rearrange.month,Rearrange.wood, label= 'Biomass-Wood')
plt.plot(Rearrange.month,Rearrange.otherbiomass, label= 'Other Biomass')

#Make the iterate of x-label
plt.xticks(Rearrange.month[::12])
plt.legend()
plt.xlabel('Month in Year')
plt.ylabel('Electricity is produced in Thousands Magewattshour')
plt.show()
```
